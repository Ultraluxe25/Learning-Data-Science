```python
 import sqlite3
```

```python
con = sqlite3.connect('db')  # database name
```

[sqlite3.connect](https://docs.python.org/3/library/sqlite3.html#sqlite3.connect)

```python
import pandas as pd
```

```python
df = pd.DataFrame({'stars': [1, 2, 3, 4, 5]})
df
```

```python
df.to_sql('test_table', con, if_exists='replace', index=False)
```

[pandas.DataFrame.to_sql](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_sql.html)

```python
sql = 'SELECT tb.stars FROM test_table tb LIMIT 3'
```

```python
pd.read_sql(sql, con)
```

[pandas.read_sql](https://pandas.pydata.org/docs/reference/api/pandas.read_sql.html?highlight=read_sql#pandas.read_sql)

```python
sql = 'SELECT tbl.stars * 100 as MANY_STARS FROM test_table tbl'
```

```python
# 'tbl' is just an alias
```

```python
result = pd.read_sql(sql, con)
result
```
